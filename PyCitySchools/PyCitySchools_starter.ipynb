```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the average reading score

```python
#Calculate the total number of schools
total_number_schools= school_data['school_name'].count()
total_number_schools
```

```python
#total number of students
total_students= student_data['Student ID'].count()
total_students
```

```python
#Calculate the total budget
total_budget= school_data['budget'].sum()
total_budget
```

```python
#Calculate the average math score (mean)
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
#Calculate the average reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
student_data['passing math'] =student_data['math_score']>=70
passing_math_percentage =student_data['passing math'].mean()*100
passing_math_percentage
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
student_data['reading score'] =student_data['reading_score']>=70
passing_reading_percentage =student_data['reading score'].mean()*100
passing_reading_percentage
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing_percentage =(passing_math_percentage + passing_reading_percentage)/2
overall_passing_percentage
```

```python
#Create a dataframe to hold the above results
district_summary_table = pd.DataFrame({"Total Number of Schools": [total_number_schools], 
            "Total Number of Students": [total_students], 
            "Total Budget": [total_budget], 
            "Average Math Score": [average_math_score], 
            "Average Reading Score": [average_reading_score], 
           "% Students with Passing Math Scores": [passing_math_percentage],
           "% Students with Passing Reading Scores": [passing_reading_percentage],
            "% Students Passing both Subjects": [overall_passing_percentage]})

#Optional: give the displayed data cleaner formatting

#round off to two decimal places
district_summary_table["Average Math Score"]=round(average_math_score,2)
district_summary_table["Average Reading Score"]=round(average_reading_score,2)
district_summary_table["% Students with Passing Math Scores"]=round(passing_math_percentage,2)
district_summary_table["% Students with Passing Reading Scores"]=round(passing_reading_percentage,2)

#add percentage sign to the required fields
district_summary_table["% Students with Passing Math Scores"] = district_summary_table["% Students with Passing Math Scores"].map("{:,.2f}%".format)
district_summary_table["% Students with Passing Reading Scores"] = district_summary_table["% Students with Passing Reading Scores"].map("{:,.2f}%".format)
district_summary_table["% Students Passing both Subjects"] = district_summary_table["% Students Passing both Subjects"].map("{:,.2f}%".format)

#add dollar sign for total budget
district_summary_table["Total Budget"] = district_summary_table["Total Budget"].map("${:,.2f}".format)

district_summary_table
```

```python
#% Passing Math

passing_math_score=school_data_complete.loc[school_data_complete["math_score"]>=70]
passing_math_score

#assign the math score by each school
group_math_per_school= passing_math_score["school_name"].value_counts()
group_math_per_school

percent_math= (group_math_per_school/total_students)*100
percent_math

```

```python
#% Passing Reading
passing_reading_score=school_data_complete.loc[school_data_complete['reading_score']>=70]
passing_reading_score

#assign the reading score by each school
reading_per_school= passing_reading_score['school_name'].value_counts()
reading_per_school

percent_reading= (reading_per_school/total_students)*100
percent_reading
```

```python
#% Overall Passing (The percentage of students that passed math and reading.)

overall_passing = (percent_math+ percent_reading)/ 2
overall_passing
```

```python
#Create a dataframe to hold the above results
school_summary = pd.DataFrame({"School Type": school_types,
                               "Total Students" : total_students,
                               "Total School Budget": school_budget,
                               "Per Student Budget":per_student_budget, 
                               "Average Math Score": average_math_score,
                               "Average Reading Score": average_reading_score, 
                               "% Passing Math": percent_math,
                               "% Passing Reading": percent_reading,
                               "% Overall Passing Rate": overall_passing})

#round off to two decimal places
school_summary["Average Math Score"]=round(average_math_score,2)
school_summary["Average Reading Score"]=round(average_reading_score,2)
school_summary["% Passing Math"]=round(percent_math,2)
school_summary["% Passing Reading"]=round(percent_reading,2)
school_summary["% Overall Passing Rate"]=round(overall_passing,2)

#Optional: give the displayed data cleaner formatting
#add dollar sign to total school budget and per student budget
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)

#display table
school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#sample bins  (establish the bins)
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"] #labels


# Categorize the spending based on the bins
school_summary["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, spending_bins, labels=group_names)

spending_math_scores = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_summary.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing Rate"]

overall_passing_rate = (spending_passing_math + spending_passing_reading) /2

#assign into a data frame
spending_summary = pd.DataFrame({"Average Math Score" : spending_math_scores,
                                 "Average Reading Score": spending_reading_scores,
                                 "% Passing Math": spending_passing_math,
                                 "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing Rate": overall_passing_rate})


# Display results
spending_summary
```
